```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
import geopy
from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter
import folium
from folium.plugins import FastMarkerCluster
```

```python
hotel = pd.read_csv('geodatabase/hotel.csv')
hotel.head(13)
```

```python
hotel.info()
```

```python
# mengubah tipe data menjadi string
hotel = hotel.astype(str)
hotel.info()
```

```python
hotel['ALAMAT'] = hotel['Alamat1'].astype(str) + ',' + \
                hotel['Alamat2'] + ',' + \
                hotel['Alamat3'] + ',' + \
                hotel['Alamat4'] + ',' + 'Indonesia' 

hotel.head(13)
```

```python
from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter

# Inisialisasi Nominatim
locator = Nominatim(user_agent="myGeocoder")

# Buat RateLimiter untuk geocoding
geocode = RateLimiter(locator.geocode, min_delay_seconds=1)

# Terapkan geocode pada kolom 'ALAMAT'
hotel['location'] = hotel['ALAMAT'].apply(geocode)

# Buat kolom 'point' berdasarkan hasil geocoding
hotel['point'] = hotel['location'].apply(lambda loc: tuple(loc.point) if loc else None)

hotel.head(13)
```
