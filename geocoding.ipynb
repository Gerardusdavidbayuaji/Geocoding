```python
import geopandas as gpd
import pandas as pd
import folium
import shutil

from geopy.extra.rate_limiter import RateLimiter
from geopy.geocoders import Nominatim
from shapely.geometry import Point
```

```python
hotel = pd.read_csv('geodatabase/hotel.csv')
hotel.head(13)
```

```python
hotel = hotel.astype(str)
hotel.info()
```

```python
hotel['Alamat'] = hotel['Alamat1'].astype(str) + ',' + \
                hotel['Alamat2'] + ',' + \
                hotel['Alamat3'] + ',' + \
                hotel['Alamat4'] + ',' + 'Indonesia' 

hotel.head(13)
```

```python
locator = Nominatim(user_agent="myGeocoder")
geocode = RateLimiter(locator.geocode, min_delay_seconds=1)
hotel['Lokasi'] = hotel['Alamat'].apply(geocode)
hotel['Koordinat'] = hotel['Lokasi'].apply(lambda loc: tuple(loc.point) if loc else None)
hotel.head(13)
```

```python
hotel[['Latitude', 'Longitude', 'Altitude']] = pd.DataFrame(hotel['Koordinat'].tolist(), index=hotel.index)
hotel.head()
```

```python
selected_columns = hotel[['Hotel', 'Alamat', 'Latitude', 'Longitude']]
selected_columns = selected_columns.dropna(subset=['Latitude', 'Longitude'])
selected_columns['geometry'] = selected_columns.apply(lambda row: Point((float(row['Longitude']), float(row['Latitude']))), axis=1)

gdf = gpd.GeoDataFrame(selected_columns, geometry='geometry')
geojson_path = 'data_hotel.geojson'
gdf.to_file(geojson_path, driver='GeoJSON')
repository_folder = 'repository/'
shutil.move(geojson_path, repository_folder)
gdf.head(13)
```

```python
selected_columns = hotel[['Hotel', 'Alamat', 'Lokasi', 'Latitude', 'Longitude']]
selected_columns = selected_columns.dropna(subset=['Latitude', 'Longitude'])

map_center = [selected_columns['Latitude'].mean(), selected_columns['Longitude'].mean()]
map = folium.Map(location=map_center, zoom_start=13)

for idx, row in selected_columns.iterrows():
    folium.Marker(
        location=[row['Latitude'], row['Longitude']],
        popup=f"Hotel: {row['Hotel']}<br>Alamat: {row['Alamat']}",
        icon=folium.Icon(icon='info-sign', color='blue') 
    ).add_to(map)

map
```
