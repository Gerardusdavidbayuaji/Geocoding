```python
import geopandas as gpd
import pandas as pd
import folium
import shutil

from geopy.extra.rate_limiter import RateLimiter
from geopy.geocoders import Nominatim
from shapely.geometry import Point
```

```python
danau = pd.read_csv('geodatabase/titik_danau_sp2.csv')
danau
```

```python
# Replace 'Not Found' values with empty strings
danau.replace('Not Found', '', inplace=True)

# Create the 'alamat' field by concatenating 'desa', 'kecamatan', and 'kabupaten'
danau['alamat'] = danau['desa'] + ', ' + danau['kecamatan'] + ', ' + danau['kabupaten'] + ', ' + danau['provinsi']

# Replace occurrences of ", ," with ", " in the 'alamat' column
danau['alamat'] = danau['alamat'].str.replace(', ,', ', ')

danau['alamat'] = danau['alamat'].str.rstrip(', ')

# Display the updated DataFrame
danau
```

```python
danau.to_excel('geodatabase/titik_danau_sp5.xlsx', index=False)
```
